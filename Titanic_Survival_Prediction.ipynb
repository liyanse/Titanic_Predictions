```python
#From the information above, as we only need to predict survival of passengers, we won't necessarily need to use data suchas
#name, ticket, cabin or even the fare paid. Thus, we'll drop these columns from our data set.
df = titanic_data.drop(['Name','Ticket','Cabin','Fare'], axis=1)
df.dropna(inplace=True)
```

```python
#The next step is to combine these different columns into one using the concept of 'one-hot'
df = pd.get_dummies(df, columns = ['Sex'])
df = pd.get_dummies(df, columns = ['Pclass'])
df = pd.get_dummies(df, columns = ['Embarked'])
df = pd.get_dummies(df, columns = ['Parch'])
df = pd.get_dummies(df, columns = ['SibSp'])
```

```python
#Let's call the method to see how our new data looks like;
df
```

```python
#With the total records above as one data frame, let's declare our independent and dependent varioables
y = df['Survived']
X = df.drop(['Survived', 'PassengerId'], axis=1)
```

```python
#For this model, I'll be using logistic regression so let's import the necessary libraries
from sklearn import linear_model
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(solver='liblinear', random_state=0)
classifier.fit(X, y)
```

```python
##If the above code runs without errors, we'll you have successfully implemented logistic regression.
#The next step, is to import the test data set
```
